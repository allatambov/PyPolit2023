### Задача 1

Напишите программу, которая запрашивает у пользователя имя и фамилию (отдельный ввод для имени и фамилии) и выводит на экран сообщение:

    Имя Фамилия, добро пожаловать!
    
**Пример работы программы:**

Ввод:

    Введите имя: Алла
    Введите фамилию: Тамбовцева
    
Вывод:

    Алла Тамбовцева, добро пожаловать!

```python
name = input("Введите имя: ")
surname = input("Введите фамилию: ")

# решение с sep
# делаем разделитель пустым, чтобы избавиться от пробелов совсем,
# но между именем и фамилией вставляем пробел, чтобы они не склеились

print(name, " ", surname, ", добро пожаловать!", sep = "") 
```

```python
# решение без sep
# склеиваем фамилию и приветствие через +,
# + просто склеивает строки и не добавляет пробелы

print(name, surname + ", добро пожаловать!") 
```

### Задача 3

Напишите программу, которая запрашивает у пользователя год рождения, информацию о том, был ли у того день рождения (`True` или пустая строка), и выводит на экран число полных лет этого пользователя.

**Пример работы программы:**

Ввод:
    
    1994
    
    
Вывод:

    28
    
Ввод:
    
    1957
    True
    
Вывод:

    66

```python
# важное замечание 1: bool() от любого текста дает True,
# bool() от пустой строки дает False (и только от нее)

bool("True")
```

```python
# неочевидно, но факт

bool("False")
```

```python
# то, что надо

bool("")
```

```python
# важное замечание 2: если True перевести в целое число, получим 1
# если False – получим 0

print(int(True), int(False))
```

```python
# запрашиваем год рождения birth
# делаем ввод типом bool – получаем либо True (ДР был), либо False (ДР не был)
# превращаем True в 1, а False в 0 через int(bool())

# вычитаем год рождения из предыдущего года,
# а затем добавляем к этим годам либо 1 (ДР был), либо 0 (ДР не был)

birth = int(input())
binary = int(bool(input()))
print(2022 -  birth + binary) 
```

### Задача 4

Напишите программу, которая запрашивает у пользователя с клавиатуры его рост в сантиметрах, его вес в килограммах (каждый показатель – с новой строки, в новом запросе) и выводит на экран сообщение вида:

    Индекс массы тела: [значение].

где вместо `[значение]` подставляется посчитанное значение индекса массы тела, округлённое до второго знака после запятой.

Индекс массы тела считается так:

$$
BMI=\frac{m}{h^2},$$

где $m$ – масса тела в килограммах, $h$ – рост *в метрах*.

**Пример работы программы:**

Ввод:
    
    Укажите рост: 168
    Укажите вес: 55
    
Вывод:
    
    Индекс массы тела: 19.49.


```python
h = int(input("Enter your height: "))
m = int(input("Enter your weight: "))

# внимание: не забываем перевести рост в метры, поделив на 100

bmi = m / (h / 100) ** 2
bmi_r = round(bmi, 2)

# если доклеиваем точку через +, превращаем bmi_r в текст string,
# так как нельзя склеить число и строку

print("Индекс массы тела:", str(bmi_r) + ".")   
```

```python
# вариант решения без +, но с sep

print("Индекс массы тела: ", bmi_r, ".", sep = "")   
```
