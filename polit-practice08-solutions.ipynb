### Задача 5 (0.5 балла)

Реализуйте программу, сохраняющую в список квадраты четных чисел.

* Задан пустой список `squares`. Этот код уже написан.
* В цикле `while` считываются целые положительные числа до тех пор, пока не введен 0 (не учитывается в последовательности).
* Для каждого вводимого числа сделайте проверку: если число четное — добавьте его квадрат в список `squares`.
* Гарантируется, что до 0 будет введено как минимум одно целое положительное число.
* Напечатайте список `squares`.

````python
squares = []

# ваш код ниже
# решение 1
while True:
    x = int(input())
    if x == 0:
        break
    elif x % 2 == 0:
        squares.append(x ** 2)
print(squares)
````

````python
squares = []

# ваш код ниже
# решение 2

x = int(input())
while x != 0:
    if x % 2 == 0:
        squares.append(x ** 2)
    x = int(input())
print(squares) 
````

В переменной `animals` хранится список:

Расставьте команды ниже таким образом, чтобы вывод программы был следующим:

    ['кошка', 'питон', 'попугайчик', 'ласка', 'тритон']

````python
animals = ['кошка', 'собака', 'питон', 'попугайчик', 'кошка']

# важно: функция sorted() умеет сортировать списки/кортежи/множества
# но на выходе всегда возвращает список
# поэтому нигде нет отдельной строки с превращением множества в список

animals = set(animals)
animals.add('тритон')
animals = sorted(animals)
animals[-2] = 'ласка'
print(animals)
````

### Задача 2 (0.5 балла)

Реализуйте программу, которая помогает двум друзьям выбрать подходящее время для совместной поездки:
    
* В переменную `months_vasya` считывается список названий месяцев (строки) — время, когда Вася может поехать в отпуск. Этот код уже написан.
* В переменную `months_dasha` считывается список названий месяцев (строки) — время, когда Даша может поехать в отпуск. Этот код уже написан.
* В переменную `bad_weather`  считывается список названий месяцев (строки) — время, когда в стране назначения плохая погода. Этот код уже написан.
* Гарантируется, что названия месяцев в каждом из трех списков уникальные и что в каждом списке есть как минимум одно название месяца.
* Выведите через запятую с пробелом названия месяцев, в которые Вася и Даша могут вместе поехать в отпуск при условии, что в месте назначения не плохая погода. Названия отсортируйте в алфавитном порядке.

**Для примера:**

Ввод

    январь март февраль ноябрь
    январь февраль июль
    ноябрь декабрь январь

Результат

    февраль

Ввод

    март
    май
    декабрь

Результат


Ввод

    июнь июль август октябрь март
    июль июнь август
    октябрь ноябрь декабрь 
    
Результат    
    
    август, июль, июнь

````python
# этот код уже дан в примере экзамена

months_vasya = input().split()
months_dasha = input().split()
bad_weather = input().split()
````

````python
# конвертируем списки в множества
# находим пересечение первых двух, 
# а затем вычитаем элементы третьего
# результат сортируем и склеиваем через .join()

vasya = set(months_vasya)
dasha = set(months_dasha)
bad = set(bad_weather)

res = vasya.intersection(dasha).difference(bad)
print(", ".join(sorted(res)))
````
