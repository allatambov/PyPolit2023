### Задача 1

В списке `flowers` хранится информация по числу собранных ромашек героями сказок С.Козлова про ёжика и медвежонка. 

Отсортируйте пары по количеству собранных ромашек, от меньшего к большему и выведите на экран записи вида:

    имя: медвежонок, ромашки: 15
    имя: ёжик, ромашки: 9
    имя: заяц, ромашки: 5
    имя: волк, ромашки: 1

```python
flowers = [("ёжик", 9), ("медвежонок", 15), 
           ("заяц", 5), ("волк", 1)]

flowers_upd = sorted(flowers, key = lambda x: x[1], reverse = True) 
for name, flower in flowers_upd:
    print(f"имя: {name}, ромашки: {flower}")
```

### Задача 2

Реализуйте программу, которая записывает в файл информацию о количестве собранных героями ромашек.

Формат ввода: 

* Строка с именами героев через пробел.
* Строка с целыми числами через пробел.

Формат вывода:

Файл `flowers.txt` с кодировкой `UTF-8`, где каждый герой, содержащий записи формате "имя: <имя>, ромашки: <число>". Каждый герой должен быть записан на отдельной строке, имена нужно отсортировать по количеству ромашек от самого большого к самому маленькому.

```python
names = input().split()
romashki = list(map(int, input().split())) 
flowers = sorted(zip(names, romashki), 
                 key = lambda x: x[1], reverse= True)

with open("flowers.txt", "w", encoding = "UTF-8") as f:
    for name, flower in flowers:
        print(f"имя: {name}, ромашки: {flower}", file = f)
```

### Задача 3

*Задача из независимого экзамена*

Реализуйте программу, которая фильтрует студентов на основе их года поступления в университет:

* В переменную `students` считывается список, где хранятся имена студентов (строки). Этот код уже написан.

* В переменную `years` считывается список, где хранятся целые числа — год начала обучения студентов в университете. Каждый год соответствует студенту из списка `students`. Этот код уже написан. 

* Гарантируется, что будет введена информация минимум для одного студента.

* В переменную `year2filter` считывается число — год поступления, который нас интересует. Этот код уже написан.

Программа выводит имена студентов, которые поступили в университет в год `year2filter` или позже, каждое на новой строке. Если таких студентов нет, то программа ничего не выводит.

**Примеры работы программы**

Ввод:

    Ангелина Сидорова, Петр Варфаломеев, Максим Кац
    2020, 2019, 2018
    2019

Вывод:

    Ангелина Сидорова
    Петр Варфаломеев

Ввод:

    Алена Леонова, Евгения Загитова
    2020, 2018
    2021
    
Вывод: (ничего не выводится)

```python
students = input().split(', ')
years = list(map(int, input().split(', ')))
year2filter = int(input())

for student, year in zip(students, years):
    if year >= year2filter:
        print(student)
```

### ЗАДАЧА C2.1

В университете проходит благотворительная лотерея. Студенты приобретают лотерейные билеты, которые вытягивают случайным образом, а через некоторое время объявляются номера выигрышных билетов. Помогите собрать информацию об участниках. 

**ФОРМАТ ВВОДА**

* Записи о приобретенных лотерейных билетах: имя покупателя и, через точку с запятой и пробел, номер билета (целое число). Каждая запись вводится с новой строки. Имя покупателя может содержать любые символы, кроме точки с запятой.

* Один студент может купить несколько билетов, но в каждой строке указывается номер только одного билета.
* Когда записи заканчиваются, на отдельной строке вводится слово "END". 


**ФОРМАТ ВЫВОДА**

* Словарь, где значения — имена студентов, а ключи — списки целых чисел (приобретенных лотерейных билетов).
* Ключи и значения должны быть добавлены в словарь в том же порядке, что и при вводе.

```python
info = {}
while True:
    inp = input()
    if inp == "END":
        break
    else:
        data = inp.split("; ")
        name = data[0]
        lottery = int(data[1]) 
        if name not in info:
            info.update({name : [lottery]}) 
        else:
            info[name].append(lottery)
print(info)
```

### ЗАДАЧА C2.2

В университете проходит благотворительная лотерея. Студенты приобретают лотерейные билеты, которые вытягивают случайным образом, а через некоторое время объявляются номера выигрышных билетов. Помогите определить, кто победил в лотерее. 

**ФОРМАТ ВВОДА**

* Записи о приобретенных лотерейных билетах: имя покупателя и, через точку с запятой и пробел, номер билета (целое число). Каждая запись вводится с новой строки. Имя покупателя может содержать любые символы, кроме точки с запятой.
* Один студент может купить несколько билетов, но в каждой строке указывается номер только одного билета.
* Когда записи заканчиваются, на отдельной строке вводится слово "END". 
* Далее в одну строку через пробел вводятся номера выигрышных билетов из приобретенных.
* Гарантируется, что был куплен хотя бы один билет. 

**ФОРМАТ ВЫВОДА**

* Имена победителей лотереи в одну строку через запятую и пробел. 
* Если один человек купил несколько выигрышных билетов, выведите его имя один раз. 
* Имена должны выводиться в алфавитном порядке.

```python
# от прошлой задачи
info = {}
while True:
    inp = input()
    if inp == "END":
        break
    else:
        data = inp.split("; ")
        name = data[0]
        lottery = int(data[1]) 
        if name not in info:
            info.update({name : [lottery]}) 
        else:
            info[name].append(lottery)
# новый код
best_list = list(map(int, input().split()))
won = []
for name, lottery in info.items():
    for lot in lottery:
        if lot in best_list:
            won.append(name)
print(", ".join(sorted(set(won))))
```
