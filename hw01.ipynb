## Домашнее задание №1

### Задание

Используя Jupyter Notebook или Google Colab, создайте ipynb-файл, который содержит следующие элементы:

* не менее двух заголовков разных уровней **(1 балл)**;
* не менее одного списка, нумерованного или ненумерованного **(1 балл)**;
* не менее одного фрагмента текста, набранного курсивом **(1 балл)**;
* не менее одного фрагмента текста, набранного полужирным шрифтом **(1 балл)**;
* не менее одной ссылки **(2 балла)**;
* не менее одного изображения **(2 балла)**;
* не менее двух ячеек с кодом; код любой, сложность кода не оценивается, однако код должен корректно запускаться **(2 балла)**.

По содержанию ipynb-файл может быть любым. Некоторые идеи:

* визитная карточка или CV;
* презентация своего домашнего питомца/хобби;
* фрагмент конспекта по какому-нибудь другому курсу (можно не совсем серьёзный).

P.S. К ячейке с кодом можно подойти творчески. Как подойти к коду творчески, если учесть, что кроме функции `print()` и вычислений мы ничего пока не разбирали? Вспомните вышивку с питоном с картинки, которую мы добавляли на занятии, и попробуйте «нарисовать» что-то из символов, вызывая `print()` несколько раз в рамках одной ячейки.

### Формат сдачи

Файлы с этим домашним заданием загружаются на Dropbox по ссылке, предложенной на странице курса. 

* Если изображение загружается по ссылке (картинка отображается в ipynb-файле, не будучи сохранённой на компьютер), в качестве решения на Dropbox можно загрузить только файл с расширением `.ipynb`. 

* Если изображение загружается из файла, сохранённого на компьютере, в качестве решения на Dropbox нужно загрузить zip-архив, в котором хранится ipynb-файл и файл с изображением. Если изображение не отображается в ipynb-файле из-за отсутствия приложенной к решению картинки, балл за этот пункт будет снят.
